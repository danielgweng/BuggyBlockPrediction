# Machine Learning Algorithm to learn characteristics of buggy blocks

## Module Imports

```python
import pickle
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = {}
with open('instr_sheharyar.pkl', 'rb') as f:
    data = pickle.load(f)

# data.keys()
# data.values()
# type(data)
instr_dict = {}
for key, val in data.items():
    instr_dict.update(val)
instr_dict.items()
```

```python
new_instr = {}
all_instr = []
for key, val in instr_dict.items():
    instructions = " ".join(val).replace(","," ")
    found = re.findall("[A-Z][A-Z]+", instructions)
    all_instr += found
    new_instr[key] = found

# new_instr.items()
print(all_instr)
```

```python
np_instr = np.array(all_instr)
unique_instr = np.unique(np_instr)
print(unique_instr)
```

```python
# instr_df = pd.DataFrame(columns=unique_instr, rows=instr_dict.keys())
instr_df = pd.DataFrame([(key, var) for (key, L) in new_instr.items() for var in L], 
                 columns=['address', 'instruction'])
# pd.DataFrame.describe(instr_df)
# pd.DataFrame(new_instr)
pd.DataFrame.describe(instr_df)
```

```python
instr_pivt = instr_df.groupby('address')['instruction'].value_counts().unstack().fillna(0)
pd.DataFrame.describe(instr_pivt)
```

```python
instr_pivt.head()
```

```python
pearsoncorr = instr_pivt.corr(method='pearson')
print(pearsoncorr)
```

```python
plt.figure(figsize = (15,15))
sb.heatmap(pearsoncorr, 
            xticklabels=pearsoncorr.columns,
            yticklabels=pearsoncorr.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)
```
